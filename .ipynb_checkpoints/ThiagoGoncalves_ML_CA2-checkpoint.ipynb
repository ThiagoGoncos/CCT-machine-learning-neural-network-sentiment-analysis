## Project Overview

This project is divided into two main parts: **Neural Network Regression** and **Semantic / Sentiment Analysis**.

The first part uses the `international-visitors-london.csv` dataset to build and compare machine learning models that predict visitor spending in London, represented by the target variable **Spend (£m)**.

The second part uses social media text data to perform sentiment analysis. The text observations are classified as **positive**, **neutral**, or **negative**, and the overall sentiment distribution is visualised using Python.

All data preparation, analysis, modelling, evaluation, visualisation, and written discussion are completed in this Jupyter Notebook using Python and Markdowns.
